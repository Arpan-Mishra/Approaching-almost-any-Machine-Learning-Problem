 images from pneumothrax dataset

```python
image_path = 'Data/train_png'
```

```python
train = pd.read_csv('Data/train.csv')
```

```python
image_ids = os.listdir(image_path)
```

```python
train_ids_1 = train['ImageId'][train['target']==1].values.tolist()
```

```python
train_ids_0 = train['ImageId'][train['target']==0].values.tolist()
```

```python
import random
i = random.randint(0, len(train_ids_1))

image_1 = Image.open(os.path.join(image_path, (train_ids_1[i] + '.png')))

i = random.randint(0, len(train_ids_0))

image_0 = Image.open(os.path.join(image_path, (train_ids_0[i] + '.png')))

plt.figure(figsize=(10,8))
plt.subplot(121)
plt.title('Pneumothrax')
plt.imshow(image_1)

plt.subplot(122)
plt.title('No Pneumothrax')
plt.imshow(image_0)

plt.show()
```
